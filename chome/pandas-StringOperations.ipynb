```python
#文字列の先頭でマッチしているか判断、抽出
monte.str.extract('([A-Za-z]+)', expand=False)
```

```python
#先頭、末尾でマッチしているか判断
monte.str.findall(r'^[^AEIOU].*[^aeiou]$')
```

| で区切られてるとき  
https://qiita.com/nkay/items/3866b3f12704ec3271ca#%E6%A8%99%E8%AD%98%E5%A4%89%E6%95%B0%E3%81%AE%E4%BD%9C%E6%88%90

```python
full_monte = pd.DataFrame({'name': monte,
                           'info': ['B|C|D', 'B|D', 'A|C',
                                    'B|D', 'B|C', 'B|C|D']})
full_monte
```

```python
full_monte['info'].str.get_dummies('|')
```

```python
s = pd.Series(['a', 'a|b', np.nan, 'a|c'], dtype="string")
s
```

```python
s.str.get_dummies(sep='|')
```
