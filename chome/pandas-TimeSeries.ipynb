時系列操作



```python
from datetime import datetime
datetime(year=2021, month=10, day=15)
```

```python
from dateutil import parser
date = parser.parse("15th of October, 2021")
date
```

```python
#曜日取得
date.strftime('%A')
```

datetime64

```python
import numpy as np
date = np.array('2021-10-15', dtype=np.datetime64)
date
```

```python
#簡単に処理できる
date + np.arange(12)
```

```python
np.datetime64('2021-10-15 12:00')
```

```python
#表示する単位を指定
np.datetime64('2021-10-15 12:59:59.50', 'ns')
```

```python
import pandas as pd
date = pd.to_datetime("15th of October, 2021")
date
```

```python
date.strftime('%A')
```

```python
#pandasで簡単に処理できる
date + pd.to_timedelta(np.arange(12), 'D')
```

時間によるインデックス

```python
index = pd.DatetimeIndex(['2020-07-04', '2020-08-04',
                          '2021-07-04', '2021-08-04'])
data = pd.Series([0, 1, 2, 3], index=index)
data
```

```python
data['2020-07-04':'2021-07-04']
```

```python
#スライス
data['2020']
```

時系列データ

```python
dates = pd.to_datetime([datetime(2021, 10, 15), '15th of October, 2021',
                       '2021-Jul-6', '07-07-2021', '20210708'])
dates
```

```python
dates.to_period('D')
```
